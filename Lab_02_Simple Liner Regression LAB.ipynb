```
%tensorflow_version 2.x
import tensorflow as tf
```

```


x_data = [1,2,3,4,5]
y_data = [2,3,4,5,6]

w = tf.Variable(61.0)
b = tf.Variable(80.0)

hx = w*x_data + b

# 요기서 minimize하는 알고리즘이 쓰인다.
# 그 중 하나가 Gradient descent.
cost = tf.reduce_mean(tf.square(hx - y_data))
print(cost)
```

```

learning_rate = 0.01


# Gradient descent 1 time.
# 반복 횟수는 Epoch!![에포크]
for i in range(1000):
  with tf.GradientTape() as tape:
    hx = w*x_data + b
    cost = tf.reduce_mean(tf.square(hx - y_data))

  W_grad, b_grad = tape.gradient(cost, [w, b])

  w.assign_sub(learning_rate * W_grad)
  b.assign_sub(learning_rate * b_grad)
  
  if i % 100 == 0:
    print("{} | {} | {} | {:0.6f}".format(i, b.numpy(), b.numpy(), cost))
    
```
